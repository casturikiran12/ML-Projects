```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
ds=pd.read_csv('sonar data.csv', header=None)
```

```python
ds.head()
```

```python
ds.shape
```

```python
ds[60].value_counts()
```

```python
X=ds.drop(columns=60,axis=1)
Y=ds[60]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#### Model prediction 

### Train data accuracry 

Y_train_predict=model.predict(X_train)

train_accuracy=accuracy_score(Y_train,Y_train_predict)

print("The accuracy of the model on training data set is :", train_accuracy)
```

```python
#### Test data accuracy 

Y_test_predict=model.predict(X_test)

test_accuracy=accuracy_score(Y_test,Y_test_predict)

print("The accuracy of the model on test data set is :", test_accuracy)
```
